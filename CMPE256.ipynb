```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g=nx.DiGraph()
g.add_node(0)
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_node(5)
g.add_edges_from([(0,1),(1,0),(1,3),(2,3),(3,4),(4,3),(4,2),(0,2),(2,5),(5,4)])

nx.draw(g)
plt.show()
```

```python
dc=nx.degree_centrality(g)
print ('Degree Centrality')
print ('%s' %dc)
```

```python
cc=nx.closeness_centrality(g)
print ('Closeness Centrality')
print ('%s' %cc)
```

```python
bc=nx.betweenness_centrality(g)
print ('Betweenness Centrality')
print ('%s' %bc)
```

```python
pg=nx.pagerank(g, alpha=0.85)
print ('Pagerank')
print ('%s' %pg)
```
